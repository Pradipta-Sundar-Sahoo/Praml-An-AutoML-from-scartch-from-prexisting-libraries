```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
from submission_repo.praml.praml_search import praml_search
from submission_repo.praml.utils import get_logger
import os
import pprint
```

```python

logger = get_logger(__name__)
pp = pprint.PrettyPrinter(indent=4) 

# Load the dataset
logger.info("Loading dataset")
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target

# Encode the target column if necessary
target_column = 'species'
le = LabelEncoder()
df_iris[target_column] = le.fit_transform(df_iris[target_column])

# Perform AutoML search with accuracy
logger.info("Starting praml search ")
result = praml_search(dataset=df_iris, target_column=target_column, problem_type="classification", basis="accuracy",max_evals=5)
```

```python
# Print the best model and its parameters
print("________________________________________________________________")
logger.info("Best classification model based on accuracy:")
pp.pprint(result.best_model)
```

```python
logger.info("Name of the best classification model:")
pp.pprint(result.best_model_name)
```

```python
print(result.best_model_name)
print(result.best_score)
```

```python
logger.info("Parameters of the best classification model:")
pp.pprint(result.parameters)
```
